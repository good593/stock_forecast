# [원-달러 환율 조회](https://kr.investing.com/currencies/usd-krw-historical-data)

```python
import pandas as pd 
```

```python
save_path_2000 = '../data/raw/USD_KRW_2000.csv'
df_2000 = pd.read_csv(save_path_2000)
df_2000 = df_2000[['날짜', '종가', '시가', '고가', '저가', '변동 %']]

df_2000.head()
```

```python
df_2000['종가'] = df_2000['종가'].map(lambda x: float(x.replace(',','')))
df_2000['시가'] = df_2000['시가'].map(lambda x: float(x.replace(',','')))
df_2000['고가'] = df_2000['고가'].map(lambda x: float(x.replace(',','')))
df_2000['저가'] = df_2000['저가'].map(lambda x: float(x.replace(',','')))
df_2000['변동 %'] = df_2000['변동 %'].map(lambda x: float(x.replace('%','')))

df_2000.head()
```

```python
save_paths = ['../data/raw/USD_KRW_2000.csv', '../data/raw/USD_KRW_2010.csv', '../data/raw/USD_KRW_2020.csv']

def get_exchange_rate(save_paths):
    df = pd.DataFrame()
    for save_path in save_paths:
        _df = pd.read_csv(save_path)
        _df = _df[['날짜', '종가', '시가', '고가', '저가', '변동 %']]
        _df['종가'] = _df['종가'].map(lambda x: float(x.replace(',','')))
        _df['시가'] = _df['시가'].map(lambda x: float(x.replace(',','')))
        _df['고가'] = _df['고가'].map(lambda x: float(x.replace(',','')))
        _df['저가'] = _df['저가'].map(lambda x: float(x.replace(',','')))
        _df['변동 %'] = _df['변동 %'].map(lambda x: float(x.replace('%','')))

        df = pd.concat([df,_df], ignore_index=True)

    return df 
```

```python
df = get_exchange_rate(save_paths)

print(f'{df.shape} / {df.isnull().sum()}')
df.head()
```

```python
save_path = '../data/exchange_rate_from_2000.csv'
df.to_csv(save_path, header=True, index=False, encoding='utf-8')
```
